Starting from the attached CSV file, I would like you to write for me a python script, executed from a jupyter notebook, that plots, using curves:
- the graph_size values in X axis
- the loading_time_seconds in Y axis
- the memory_used_mb in the Y axis

Please make sure that:
-  the labels for loading_time_seconds is on the left Y axis
-  the labels for the memory_used_mb is on the right Y axis
  

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '../out/loading-metrics-20250328-171639.csv'
data = pd.read_csv(file_path)

# Plotting the graph
plt.figure(figsize=(10, 6))

# Plotting loading_time_seconds
plt.plot(data['graph_size'], data['loading_time_seconds'], label='Loading Time (s)', marker='o')

# Creating a secondary y-axis for memory_used_mb
plt.gca().twinx().plot(data['graph_size'], data['memory_used_mb'], label='Memory Used (MB)', color='r', marker='x')

# Adding labels and title
plt.xlabel('Graph Size')
plt.ylabel('Loading Time (s) / Memory Used (MB)')
plt.title('Loading Time and Memory Usage vs Graph Size')

# Adding legends
plt.legend(loc='upper left')
plt.gca().twinx().legend(loc='upper right')

# Display the plot
plt.show()
```
