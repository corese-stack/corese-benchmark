Hi Chat

Starting from the attached CSV file, I would like you to write for me a python script, executed from a jupyter notebook, that plots, using curves:
- the graph_size values in X axis
- the loading_time_seconds in Y axis
- the memory_used_mb in the Y axis

Please make sure that:
-  the labels for loading_time_seconds is on the left Y axis
-  the labels for the memory_used_mb is on the right Y axis
-  the values for X axis are shown as millions (10^6) multiples. 
  

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
#file_path = '../out/rdf4j.5.1.2_loading-metrics.csv'
file_path = '../out/corese.4.6.3_loading-metrics.csv'
data = pd.read_csv(file_path)

# Convert graph_size to millions for better readability
data['graph_size_millions'] = data['graph_size'] / 1e6

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot loading_time_seconds on the left Y-axis
ax1.set_xlabel('Graph Size (Millions)')
ax1.set_ylabel('Loading Time (Seconds)', color='tab:blue')
ax1.plot(data['graph_size_millions'], data['loading_time_seconds'], color='tab:blue', marker='o', label='Loading Time')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second Y-axis for memory_used_mb
ax2 = ax1.twinx()
ax2.set_ylabel('Memory Used (MB)', color='tab:red')
ax2.plot(data['graph_size_millions'], data['memory_used_mb'], color='tab:red', marker='x', label='Memory Used', linestyle='--')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add a title
plt.title('Loading Metrics vs Graph Size')

# Show the plot
plt.show()
```
